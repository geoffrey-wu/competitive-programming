```python
def flood(grid, i, j, visited=None, sides=None):
    visited=visited or set()
    sides = sides or set()
    if (i, j) in visited:
        return 0, 0
    visited.add((i, j))
    x = grid[i][j]
    a = 1
    p = 4
    if i > 0 and grid[i-1][j] == grid[i][j]:
        r = flood(grid, i-1, j, visited, sides)
        a += r[0]
        p += r[1] - 1
    else:
        sides.add((i, j, 'top'))
    if i < len(grid)-1 and grid[i+1][j] == grid[i][j]:
        r = flood(grid, i+1, j, visited, sides)
        a += r[0]
        p += r[1] - 1
    else:
        sides.add((i, j, 'bottom'))
    if j > 0 and grid[i][j-1] == grid[i][j]:
        r = flood(grid, i, j-1, visited, sides)
        a += r[0]
        p += r[1] - 1
    else:
        sides.add((i, j, 'left'))
    if j < len(grid[i])-1 and grid[i][j+1] == grid[i][j]:
        r = flood(grid, i, j+1, visited, sides)
        a += r[0]
        p += r[1] - 1
    else:
        sides.add((i, j, 'right'))
    return a, p, visited, sides
```

```python
def edges(s):
    e = 0
    while len(s):
        a = s.pop()
        if a[2] == 'top' or a[2] == 'bottom':
            i = a[1] + 1
            while (a[0], i, a[2]) in s:
                s.remove((a[0], i, a[2]))
                i += 1
            i = a[1] - 1
            while (a[0], i, a[2]) in s:
                s.remove((a[0], i, a[2]))
                i -= 1
        else:
            i = a[0] + 1
            while (i, a[1], a[2]) in s:
                s.remove((i, a[1], a[2]))
                i += 1
            i = a[0] - 1
            while (i, a[1], a[2]) in s:
                s.remove((i, a[1], a[2]))
                i -= 1
        e += 1
    return e
```

```python
visited = set()
total = 0
for i in range(len(grid)):
    for j in range(len(grid[i])):
        if (i, j) not in visited:
            r = flood(grid, i, j)
            e = edges(r[3])
            visited |= r[2]
            total += r[0]*e
print(total)
```
